```python
#if you need an API Key from OpenAI
#https://platform.openai.com/account/api-keys

from openai import OpenAI
import os
from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv())

OPENAI_API_KEY  = os.getenv('OPENAI_API_KEY')
```

```python
client = OpenAI(
    # This is the default and can be omitted
    api_key=OPENAI_API_KEY,
)

def continue_conversation(messages, temperature=0):
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=messages,
        temperature=temperature,
    )
    #print(str(response.choices[0].message["content"]))
    return response.choices[0].message.content
```

```python
def add_prompts_conversation(_):
    #Get the value introduced by the user
    prompt = client_prompt.value_input
    client_prompt.value = ''

    #Append to the context the User prompt.
    context.append({'role':'user', 'content':f"{prompt}"})

    #Get the response.
    response = continue_conversation(context)

    #Add the response to the context.
    context.append({'role':'assistant', 'content':f"{response}"})

    #Update the panels to show the conversation.
    panels.append(
        pn.Row('User:', pn.pane.Markdown(prompt, width=600)))
    panels.append(
        pn.Row('Assistant:', pn.pane.Markdown(response, width=600)))

    return pn.Column(*panels)
```

   - Each of the three chatbot personalities—Normal, Friendly, and Effective—meets different user needs. The Normal bot is friendly and balanced, ideal for general use, but lacks personality. The Friendly bot is friendly and interactive, making it ideal for building good relationships, but it can slow down interactions. The Effective bot prioritizes speed and clarity and is ideal for quick requests, although it can appear impersonal.

```python

# three different chatbot
prompts = {
    "normal": """
    You are an OrderBot for a fast-food restaurant called 'My Dear Frankfurt'.
    Greet the customer politely, collect their order, and confirm the details with them before finalizing the order.
    Clarify any options, extras, and sizes as needed to ensure the customer is satisfied with their selection.
    Provide a summary of the order and ask if they would like to add anything else before proceeding with the payment.
    Respond in a concise and polite manner.
    """,

    "friendly": """
    Hey there! You're an OrderBot working at 'My Dear Frankfurt' – a super cool fast-food joint.
    Greet the customer with excitement, ask for their order, and help them choose their favorite meal.
    Have fun with the conversation – crack some jokes and suggest popular items like burgers, fries, and drinks.
    Confirm the order and ask if they want to add any extras before finalizing it.
    Keep it light-hearted, friendly, and fun!
    """,

    "efficient": """
    You are a fast-food OrderBot working at 'My Dear Frankfurt'.
    Greet the customer, ask for their order, and get all the details quickly without unnecessary chit-chat.
    Focus on essential information only – clarify options, extras, and sizes, then provide a quick summary of the order.
    Confirm the final order and move on to the payment swiftly.
    Keep responses short and direct.
    """
}

# Choose a prompt style: "normal", "friendly", or "efficient"
prompt_style = "normal"

# Function to get completion from OpenAI
def get_completion_from_messages(messages, model="gpt-3.5-turbo", temperature=0):
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=temperature,
    )
    return response.choices[0].message.content

# Function to handle conversation
def collect_messages(_):
    prompt = inp.value  # Get user input
    inp.value = ''  # Clear input field
    context.append({'role': 'user', 'content': prompt})  # Store user message

    response = get_completion_from_messages(context)  # Get GPT response
    context.append({'role': 'assistant', 'content': response})  # Store response

    panels.append(pn.Row("User:", pn.pane.Markdown(prompt, width=600)))
    panels.append(pn.Row("Assistant:", pn.pane.Markdown(response, width=600, styles={'background-color': '#F6F6F6'})))

    return pn.Column(*panels)

pn.extension()

panels = []

# Set chatbot context using selected prompt style
context = [
    {'role': 'system', 'content': prompts[prompt_style]}
]

# UI Elements
inp = pn.widgets.TextInput(value="", placeholder='Enter text here…')
button_conversation = pn.widgets.Button(name="Chat!")

interactive_conversation = pn.bind(collect_messages, button_conversation)

dashboard = pn.Column(
    inp,
    pn.Row(button_conversation),
    pn.panel(interactive_conversation, loading_indicator=True, height=300),
)

dashboard
```
